#Regular Expression

```python
import nltk
nltk.download('brown')
from nltk.corpus import brown
```

```python
import re
```

```python
brown_corpus = brown.sents()
all_words = [word.lower() for sentence in brown_corpus for word in sentence]
pattern = r'\bthe\b'
```

```python
instances_of_the = [word for word in all_words if re.match(pattern,word)]
```

```python
print("instances:",instances_of_the)
```

```python
count_the = len(instances_of_the)
print("no. of instances of the is:",count_the)
```

#Find Bigrams and Trigrams

```python
from nltk import bigrams, trigrams
```

```python
all_bigrams = list(bigrams(all_words))
all_trigrams = list(trigrams(all_words))
```

```python
print("Generated bigrams:")
print(all_bigrams[:10])
```

```python
print("Generated trigrams:")
print(all_trigrams[:10])
```

#Word Tokenizer

```python
!pip install pymupdf
```

```python
import fitz
```

```python
pdf_path = "/content/H0Mymr3OkP.pdf"
```

```python
def extract_text_from_pdf(pdf_path):
  doc = fitz.open(pdf_path)
  text = ""
  for page_num in range(doc.page_count):
    page = doc.load_page(page_num)
    text += page.get_text()
  doc.close()
  return text

pdf_text = extract_text_from_pdf(pdf_path)
```

```python
tokens = word_tokenize(pdf_text)
print(tokens)
```

#Whitespace Tokenizer

```python
whitespace_tokens = pdf_text.split()
print(whitespace_tokens)
```

#Sentence Tokenizer

```python
from nltk.tokenize import PunktSentenceTokenizer
```

```python
tokenizer = PunktSentenceTokenizer()
tokenizer.train(pdf_text)

punkt_tokens = tokenizer.tokenize(pdf_text)
```

```python
for index, sentence in enumerate(punkt_tokens, start = 1):
  print(f"Sentence {index}: {sentence}\n")
```
